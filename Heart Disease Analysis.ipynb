# 1️⃣ Import Libraries and Load Dataset


### Objective: Import necessary libraries (pandas, matplotlib.pyplot, seaborn) and load the dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv(r'C:\Users\heart.csv')
data.head()
```

# 2️⃣ Check for Duplicate Data


### Objective: Identify and remove duplicates.



```python
# Check for duplicates
duplicate_count = data.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")

# Remove duplicates if any
data = data.drop_duplicates()
```

```python
data.shape
```

```python
print("Number of rows: ", data.shape[0])
print("Number of columns: ", data.shape[1])
```

```python
data.info()
```

# 3️⃣ Correlation Matrix


### Objective: Visualize correlations among features to identify strong relationships.


```python
data.describe()
```

```python
plt.figure(figsize=(10, 5))
sns.heatmap(data.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

# 4️⃣ Heart Disease Incidence


### Objective: Analyze the target variable to see how many individuals have heart disease.


```python
sns.countplot(x='target', data=data)
plt.title("Heart Disease Incidence")
plt.xlabel("Heart Disease (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

# 5️⃣ Gender Distribution


### Objective: Check the distribution of genders.


```python
sns.countplot(x='sex', data=data)
plt.title("Gender Distribution")
plt.xlabel("Gender (0 = Female, 1 = Male)")
plt.ylabel("Count")
plt.show()
```

# 6️⃣ Gender Distribution by Heart Disease Incidence


### Objective: Compare the gender distribution for individuals with and without heart disease.

```python
sns.countplot(x='sex', hue='target', data=data)
plt.title("Gender Distribution by Heart Disease Incidence")
plt.xticks([1,0],['Male','Female'])
plt.ylabel("Count")
plt.legend(title="Heart Disease")
plt.show()
```

# 7️⃣ Age Distribution


### Objective: Visualize the age range in the dataset.



```python
plt.hist(data['age'], bins=20, edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

# 8️⃣ Chest Pain Type Analysis


### Objective: Analyze the distribution of chest pain types.



```python
sns.countplot(x='cp', data=data)
plt.title("Chest Pain Type Distribution")
plt.xlabel("Chest Pain Type")
plt.ylabel("Count")
plt.show()
```

# 9️⃣ Chest Pain by Heart Disease Incidence


### Objective: Analyze chest pain types relative to heart disease.



```python
sns.countplot(x='cp', hue='target', data=data)
plt.title("Chest Pain by Heart Disease Incidence")
plt.xlabel("Chest Pain Type")
plt.ylabel("Count")
plt.legend(title="Heart Disease")
plt.show()
```

# 🔟 Fasting Blood Sugar Distribution


### Objective: Explore fasting blood sugar levels

```python
sns.countplot(x='fbs', hue='target', data=data)
plt.title("Fasting Blood Sugar Distribution by Heart Disease Incidence")
plt.xlabel("Fasting Blood Sugar > 120 mg/dl (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.legend(title="Heart Disease")
plt.show()
```

# 1️⃣1️⃣ Resting Blood Pressure Distribution


### Objective: Visualize resting blood pressure values.



```python
plt.hist(data['trestbps'], bins=15, edgecolor='black')
plt.title("Resting Blood Pressure Distribution")
plt.xlabel("Resting Blood Pressure")
plt.ylabel("Count")
plt.show()
```

# 1️⃣2️⃣ Resting Blood Pressure by Gender


### Objective: Compare resting blood pressure levels by gender.


```python
sns.boxplot(x='sex', y='trestbps', data=data)
plt.title("Resting Blood Pressure by Gender")
plt.xlabel("Gender (0 = Female, 1 = Male)")
plt.ylabel("Resting Blood Pressure")
plt.show()
```

# 1️⃣3️⃣ Serum Cholesterol Distribution


### Objective: Explore serum cholesterol levels.


```python
plt.hist(data['chol'], bins=20, edgecolor='black')
plt.title("Serum Cholesterol Distribution")
plt.xlabel("Serum Cholesterol (mg/dl)")
plt.ylabel("Count")
plt.show()
```

# 1️⃣4️⃣ Plot Continuous Variables


### Objective: Plot histograms for continuous variables to see their distributions.



```python
continuous_vars = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
data[continuous_vars].hist(bins=15, edgecolor='black', figsize=(15, 10))
plt.suptitle("Distribution of Continuous Variables")
plt.show()
```
