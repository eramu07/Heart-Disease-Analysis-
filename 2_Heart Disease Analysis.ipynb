#### üìç Importing Libraries and Dataset

##### We start by importing essential libraries and loading the dataset:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv(r'C:\Users\wasee\OneDrive\Desktop\Project\heart.csv')
df.head()
```

### üìç Cleaning Data: Checking and Dropping Duplicates
#### Data cleaning ensures reliability. Let's check and remove duplicate records:

```python
# Check for duplicates
print(f"Duplicate records: {df.duplicated().sum()}")

# Drop duplicates
df = df.drop_duplicates()
```

### üìç Correlation Matrix: Unveiling Relationships
#### A heatmap helps us visualize the correlation between features:

```python
# Correlation matrix
plt.figure(figsize=(8, 5))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

### üìç Analyzing Heart Disease Incidence
#### Using a count plot, we compare the number of individuals with and without heart disease:

```python
# Count plot for target variable
sns.countplot(x ="target", data=df, palette ="Set2")
plt.title("Heart Disease Incidence")
plt.xlabel("Heart Disease (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

### üìç Gender Distribution
#### Let's explore gender distribution:

```python
# Gender distribution
sns.countplot(x="sex", data=df, palette ="Set2")
plt.title("Gender Distribution")
plt.xlabel("Sex (0 = Female, 1 = Male)")
plt.ylabel("Count")
plt.show()
```

### üìç Gender Distribution According to Target Variable



```python
# Gender distribution by target variable
sns.countplot(x="target", hue="sex", data=df, palette="muted")
plt.title("Gender Distribution by Heart Disease")
plt.xlabel("Heart Disease (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.legend(["Female", "Male"])
plt.show()
```

### üìç Age Distribution
#### A histogram gives insights into the age range of individuals:

```python
# Age distribution
df["age"].plot(kind="hist", bins=20, color="skyblue", edgecolor="black")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

### üìç Chest Pain Type Analysis
#### Chest pain is categorized into four types:

```python
# Chest pain type count
sns.countplot(x="cp", data=df, palette="viridis")
plt.title("Chest Pain Type Distribution")
plt.xlabel("Chest Pain Type (0 to 3)")
plt.ylabel("Count")
plt.show()
```

### üìç Chest Pain Distribution by Target Variable

```python
# Chest pain by target variable
sns.countplot(x="cp", hue="target", data=df)
plt.title("Chest Pain Type vs. Heart Disease")
plt.xlabel("Chest Pain Type")
plt.ylabel("Count")
plt.legend(["No Heart Disease", "Heart Disease"])
plt.show()
```

### üìç Fasting Blood Sugar Distribution

```python
# Fasting blood sugar distribution
sns.countplot(x="fbs", hue="target", data=df, palette="husl")
plt.title("Fasting Blood Sugar by Heart Disease")
plt.xlabel("FBS (0 = <120 mg/dl, 1 = >120 mg/dl)")
plt.ylabel("Count")
plt.show()
```

### üìç Serum Cholesterol Distribution



```python
# Serum cholesterol distribution
df["chol"].plot(kind="hist", bins=30, color="orange", edgecolor="black")
plt.title("Serum Cholesterol Distribution")
plt.xlabel("Cholesterol")
plt.ylabel("Frequency")
plt.show()
```

### üìç Visualizing Continuous Variables



```python
# Continuous variables distribution
continuous_vars = ["age", "trestbps", "chol", "thalach"]
df[continuous_vars].hist(bins=20, figsize=(10, 8), color="skyblue", edgecolor="black")
plt.suptitle("Continuous Variables Distribution")
plt.show()
```
